```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cwd = os.getcwd()

print(f'Current working directory: {cwd}')
```

```python
dataset_path = os.path.join(cwd,'Datasets')

for root,dirs,filenames in os.walk(dataset_path):
    for filename in filenames:
        dataset_path = os.path.join(dataset_path, filename)
print(dataset_path)
```

```python
df = pd.read_csv(dataset_path)
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

Here time, length,dayofweek and class is the numerical data
on the other hand Airline, airportfrom, airportto is the categorical data


```python
print(f'Minimum time: {df["Time"].min()}')
print(f'Maximum time: {df["Time"].max()}')
```

in a day there are 1440 min. So, 1439 means 23:59

```python
#set Flight column as index
df.set_index('Flight',inplace=True)
df.head()
```

```python
#change categorical columns to be a category data type
for col in df.columns:
    if col in ['Time','Length','Class','DayOfWeek']:
        next
    else:
        df[col] = df[col].astype('category')
df.dtypes
```

```python
# this actually showed distinct values within that column and counts how many times it appear
df['Airline'].value_counts()
```

```python
# df['Airline'].value_counts().index
df['Airline'].value_counts(ascending=True).index
```

```python
df['Airline'].value_counts(ascending=True).values
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()

# correlation only works with numerical data
```

```python
sns.countplot(data=df, x=df['Airline'], order=df['Airline'].value_counts().index).set(title='How much flights are operated by the airlines')
plt.show()
```

```python
plt.pie(x=df['Airline'].value_counts(), labels=df['Airline'].value_counts().index, autopct='%1.1f%%',radius=2)
plt.show()
```

```python
df.columns
```

```python
df.groupby(['Airline', 'Class']).count()
```

```python
# compute the size of each group

df.groupby(['Airline', 'Class']).size()
```

```python
# want to know specific column's aggregation using goupby
df.groupby(['Airline', 'Class'])[["Length"]].agg(['size','sum', 'mean', 'max','min'])
```

```python
sns.countplot(data=df, x='DayOfWeek',hue = "Class").set(title='Total flights (not delayed and delayed) in 7 days')
plt.legend(labels = ['Not Delayed (0)', 'Delayed (1)'])
plt.show()
```

```python
# how many flights are delayed by airlines

delayed_count = df.groupby('Airline')['Class'].sum().sort_values(ascending=False)

delayed_count
```

```python
delayed_count.plot(kind='bar')
plt.ylabel('Number of delayed flights')
plt.title('Number of delayed flights per airline')
plt.show()
```
